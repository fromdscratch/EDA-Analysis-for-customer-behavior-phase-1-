```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from mlxtend.frequent_patterns import apriori, association_rules

# Load the dataset
df = pd.read_csv('/kaggle/input/store-transaction-data/Hackathon_Ideal_Data.csv')

# Data Preprocessing
# Assume no missing values for simplicity
# Convert categorical variables to appropriate data types if necessary

# Exploratory Data Analysis (EDA)
plt.figure(figsize=(10, 6))
sns.countplot(x='MONTH', data=df)
plt.title('Transaction Count by Month')
plt.xlabel('Month')
plt.ylabel('Transaction Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(x='QTY', data=df, bins=30)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='MONTH', y='VALUE', data=df)
plt.title('Total Value Over Time')
plt.xlabel('Month')
plt.ylabel('Total Value')
plt.show()

# Customer Segmentation using K-means clustering
# Assuming relevant features are available
# Example: clustering based on 'QTY' and 'VALUE'
X = df[['QTY', 'VALUE']]
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='QTY', y='VALUE', hue='Cluster', data=df, palette='Set2')
plt.title('Customer Segmentation')
plt.xlabel('Quantity')
plt.ylabel('Value')
plt.legend(title='Cluster')
plt.show()

# Save the clustered data to a CSV file
df.to_csv('Hackathon_Ideal_Data_Cluster.csv', index=False)
```
